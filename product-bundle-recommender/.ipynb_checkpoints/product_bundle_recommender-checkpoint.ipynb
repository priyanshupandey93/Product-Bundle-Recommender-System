### Import Libraries in Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

## Data Preprocessing

### Loading Dataset

```python
df = pd.read_csv(r"C:\DS&AI\Product Bundle Recommender Project\product-bundle-recommender\data.csv")
```

## Cleaning the Data
#### Remove missing/null rows
#### Drop rows with "Order ID" or "Product" as NaN
#### Filter out invalid headers inside the data (some files repeat headers)

```python
df.dropna(inplace=True)
df = df[df['Order ID'].str[0:1].str.isnumeric()]
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
```

```python
#Drop rows again if any NaNs introduced during conversion
df.dropna(inplace=True)
```

```python
df['Total Price'] = df['Quantity Ordered'] * df['Price Each']
```

```python
# Reset index
df.reset_index(drop=True, inplace=True)
```

## Frequently Bought Together Products (Market Basket Analysis)
##### We’ll use the Apriori algorithm to find combinations of products often bought together.

### Creating Transactions

#### Group Products by Order ID

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df_grouped = df[['Order ID', 'Grouped']].drop_duplicates()
```

#### Convert to List of Transactions

```python
transactions = df_grouped['Grouped'].apply(lambda x: x.split(',')).tolist()
```

### Market Basket Analysis

#### Convert to One-Hot Encoded Format

```python
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_array, columns=te.columns_)
```

### Applying Apriori Algorithm

```python
frequent_itemsets = apriori(df_encoded, min_support=0.002, use_colnames=True)
```

### Generating Association Rules

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.05)
rules = rules.sort_values(['confidence', 'lift'], ascending=[False, False])
```

## Visualizing Product Bundles (Association Rules)
##### We'll use a network graph and a bar chart for the top product pairs.

### 1. Bar Chart – Top 10 Recommended Product Pairs

```python
import seaborn as sns

# Sort rules by lift and select top 10
top_rules = rules.sort_values(by='lift', ascending=False).head(10)

# Create a column for product pair
top_rules['Product Pair'] = top_rules['antecedents'].apply(lambda x: ', '.join(list(x))) + " ➡️ " + \
                            top_rules['consequents'].apply(lambda x: ', '.join(list(x)))

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x='lift', hue='Product Pair', data=top_rules, palette='viridis')
plt.title('Top 10 Product Bundle Recommendations (by Lift)')
plt.xlabel('Lift')
plt.ylabel('Product Bundles')
plt.tight_layout()
plt.show()
```

### 2. Network Graph – Visualize Relationships

```python
import networkx as nx

# Take top 5 rules for clear visualization
top_5 = top_rules.head(5)

# Create graph
G = nx.DiGraph()

for _, row in top_5.iterrows():
    for antecedent in row['antecedents']:
        for consequent in row['consequents']:
            G.add_edge(antecedent, consequent, weight=row['lift'])

# Draw graph
plt.figure(figsize=(10,6))
pos = nx.spring_layout(G, k=0.5)
edges = G.edges(data=True)

nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=12, arrows=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): f"{d['weight']:.2f}" for u, v, d in edges})
plt.title("Top Product Bundles – Network View")
plt.show()
```

## Customer Segmentation

### 1. Extract Customer Info

```python
# Extract customer data from 'Purchase Address' as a proxy (assuming each address = unique customer)
df['Customer ID'] = df['Purchase Address']

# Calculate total spending and order count for each customer
customer_summary = df.groupby('Customer ID').agg({
    'Sales': 'sum',
    'Order ID': 'nunique',
    'City': 'first'
}).reset_index()

customer_summary.rename(columns={'Sales': 'Total_Spending', 'Order ID': 'Total_Orders'}, inplace=True)

customer_summary.head()
```

### 2. Create Segments

```python
def segment_customer(row):
    if row['Total_Spending'] > 2000:
        return 'Premium Buyer'
    elif row['Total_Orders'] > 20:
        return 'Frequent Buyer'
    elif row['Total_Spending'] < 500:
        return 'Budget Buyer'
    else:
        return 'Standard Buyer'

customer_summary['Segment'] = customer_summary.apply(segment_customer, axis=1)

customer_summary.head()
```

### 3. Visualize Segments

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.countplot(data=customer_summary, x='Segment', hue='Segment', palette='viridis', legend=False)
plt.title('Customer Segmentation Based on Spending & Orders')
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
plt.show()
```

## Building Bundle Recommender

#### Recommending Products Based on Rules

```python
def recommend_bundle(product, rules):
    recommendations = []
    for _, rule in rules.iterrows():
        if product in rule['antecedents']:
            rec = {
                'Buy With': ', '.join(rule['consequents']),
                'Confidence': round(rule['confidence'], 2),
                'Lift': round(rule['lift'], 2)
            }
            recommendations.append(rec)
    
    if recommendations:
        return recommendations
    else:
        return f"No strong bundle recommendations found for '{product}'"
```
